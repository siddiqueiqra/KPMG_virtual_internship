```python
def age_calculation(dob):
    dob=pd.to_datetime(dob)
    today=date.today()
    return (today.year-3) - dob.year
new_df['Age'] = new_df['DOB'].apply(age_calculation)
print(set(new_df["Age"].to_list()))
```

```python
# Create age ranges
bins = [0, 20, 30, 40, 50, 60 ,70,80, np.inf]
labels = ['0-19', '20-29', '30-39', '40-49', '50-59', '60-69','70-79','80-90']

# Assign each customer to an age group
new_df['age_group'] = pd.cut(new_df['Age'], bins, labels=labels)
```

```python
new_df['Profit']= new_df['list_price']-new_df['standard_cost']
```

# Recency ,Frequency and monetary value Analysis

```python
new_df.columns
```

> ### Calculate the Recency, Frequency, and Monetary Value of each customer

```python
new_df['transaction_date']=pd.to_datetime(new_df['transaction_date'])
```

```python
max_date=new_df['transaction_date'].max()
```

```python
max_date
```

```python
new_df['Recency']=(max_date - new_df['transaction_date']).dt.days
```

```python
newdf_grouped = new_df.groupby(['customer_id']).agg({
    'product_id':'count',
    'Profit':'sum'
})
```

```python
newdf_grouped.rename(columns={'product_id':'Frequency',
                             'Profit':'Monetary Value',
                             'customer_id':'Recency'},inplace=True)
```

> Assigning each customer a score for each of the RFM metrics

```python
r_quartile =pd.qcut(new_df['Recency'],4,labels=[4,3,2,1])
f_quartile = pd.qcut(newdf_grouped['Frequency'],4,labels=[1,2,3,4])
m_quartile =pd.qcut(newdf_grouped['Monetary Value'],4,labels=[1,2,3,4])
newdf_grouped['R']=r_quartile
newdf_grouped['F']=f_quartile
newdf_grouped['M']=m_quartile
```

```python
newdf_grouped['RFM_Score'] = newdf_grouped[['R', 'F', 'M']].sum(axis=1)
```

```python
def rfm_level(newdf_grouped):
    if newdf_grouped['RFM_Score']>=10:
        return 'High Value Customer'
    elif newdf_grouped['RFM_Score']>=10:
        return 'High Value Customer' 
    elif newdf_grouped['RFM_Score']>=6:
        return 'Mid Value Customer'
    else:
        return 'Low Value Customer'
```

```python
newdf_grouped['RFM_Level'] = newdf_grouped.apply(rfm_level, axis=1)
```
